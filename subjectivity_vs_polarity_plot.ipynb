

### Plot: Polarity and subjectivity
Plot the polarity and subjectivity as a scatter plot. 

Polarity: Tells us how positive is our tweet or how negative is our tweet statements

Subjectivity : It is of many forms, which tells personal opinion, expresses personal belief, views, feelings, allegations,speculations, suspicions, desires etc.
Subjective sentences generally refer to personal opinion, emotion or judgment whereas objective refers to factual information. Subjectivity is also a float which lies in the range of [0,1].



### Analyzing raw tweets from data/raw_tweets_mar_1_5.csv  (Between 1st March to 5th March 2020)

```python

import matplotlib.pyplot as plt 
import pandas as pd
import requests
from io import StringIO
from pylab import figure, axes, pie, title, show

# Create a dataframe with a column called Tweets
#df = pd.DataFrame([tweet.full_text for tweet in posts], columns=['Tweets'])
# Show the first 5 rows of data


df = pd.read_csv(r"data/market_mar_1_5.csv", encoding ="utf-8")
#print(df.head())
#print(df.shape[0])
#print(df["polarity"][2])


#for val in df.Text: 

#Plotting 
plt.figure(figsize=(8,6)) 
for i in range(0, df.shape[0]):
  plt.scatter(df["polarity"][i], df["subjectivity"][i], color='Blue') 
# plt.scatter(x,y,color)   

#with grid plot
plt.grid(True)


plt.title('Sentiment Analysis') 
plt.xlabel('Polarity') 
plt.ylabel('Subjectivity') 
plt.savefig('result_plots/Subjectivity_vs_Polarity_market_mar_1_5.png')
plt.show()
```

It looks like the tweets are equidistributed in positive,neutral and negative sentiments
This is because as there are many of the points which are on both left and right side of the polarity at value 0.
